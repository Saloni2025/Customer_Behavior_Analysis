```python
#create a column age_group 
labels=["Young Adult","Adult","Middle-aged","Senior"]
#qcut:=split the ages in 4 equal size groups based on data distribuations and assign the lables that we are defined
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
#wrok with text is harder because to use in matrix and comparing purpose it make difficult so we create new column
#create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
#we create dictionary known as frequency_mapping
#maps each textual frequency equvalient number of days
df['purchase_frequency_days'] =  df['frequency_of_purchases'].map(frequency_mapping)
```
